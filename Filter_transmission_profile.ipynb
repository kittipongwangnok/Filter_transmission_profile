```python
u_band = open("u-band_no2.txt",'r').readlines()
N_u_band = len(u_band)

#Read data
u_Wavelength = []
u_Transmission = []
#Please change the input file
for line in open("u-band_no2.txt"):
    li=line.strip()
    if not li.startswith("#"):
        u_Wavelength.append(float(li.split("\t\t")[0]))
        u_Transmission.append(float(li.split("\t\t")[1]))

        
g_band = open("g-band_no2.txt",'r').readlines()
N_g_band = len(g_band)
#Read data
g_Wavelength = []
g_Transmission = []
#Please change the input file
for line in open("g-band_no2.txt"):
    li=line.strip()
    if not li.startswith("#"):
        g_Wavelength.append(float(li.split("\t\t")[0]))
        g_Transmission.append(float(li.split("\t\t")[1]))
        
r_band = open("r-band_no2.txt",'r').readlines()
N_r_band = len(r_band)
#Read data
r_Wavelength = []
r_Transmission = []
#Please change the input file
for line in open("r-band_no2.txt"):
    li=line.strip()
    if not li.startswith("#"):
        r_Wavelength.append(float(li.split("\t\t")[0]))
        r_Transmission.append(float(li.split("\t\t")[1]))
        
i_band = open("i-band_no2.txt",'r').readlines()
N_i_band = len(i_band)
#Read data
i_Wavelength = []
i_Transmission = []
#Please change the input file
for line in open("i-band_no2.txt"):
    li=line.strip()
    if not li.startswith("#"):
        i_Wavelength.append(float(li.split("\t\t")[0]))
        i_Transmission.append(float(li.split("\t\t")[1]))
        
        
z_band = open("z-band_no2.txt",'r').readlines()
N_z_band = len(z_band)
#Read data
z_Wavelength = []
z_Transmission = []
#Please change the input file
for line in open("z-band_no2.txt"):
    li=line.strip()
    if not li.startswith("#"):
        z_Wavelength.append(float(li.split("\t\t")[0]))
        z_Transmission.append(float(li.split("\t\t")[1]))
        
kg5_band = open("kg5_band.txt",'r').readlines()
N_kg5_band = len(kg5_band)
#Read data
kg5_Wavelength = []
kg5_Transmission = []
kg5_Transmission_100 = []
#Please change the input file
for line in open("kg5_band.txt"):
    li=line.strip()
    if not li.startswith("#"):
        kg5_Wavelength.append(float(li.split("\t")[0]))
        kg5_Transmission.append(float(li.split("\t")[1]))
        kg5_Transmission_100.append(float(li.split("\t")[2]))
```

```python
##Plotgraph
fig=plt.figure(figsize=(9, 6), tight_layout=True)
plt.tick_params(direction='in', which='both', bottom='on',top='on', right = 'on')

plt.plot(u_Wavelength, u_Transmission, color = 'purple', label = 'SDSS $u^\prime$')
plt.plot(g_Wavelength, g_Transmission, color = 'blue', label = 'SDSS $g^\prime$')
plt.plot(r_Wavelength, r_Transmission, color = 'green', label = 'SDSS $r^\prime$')
plt.plot(i_Wavelength, i_Transmission, color = 'orange', label = 'SDSS $i^\prime$')
plt.plot(z_Wavelength, z_Transmission, color = 'red', label = 'SDSS $z^\prime$')
plt.plot(kg5_Wavelength, kg5_Transmission_100, color = 'magenta', label = 'Schott KG5')

plt.xlim(250,1100)
plt.ylim(0,100)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Transmission ($\%$)')

plt.legend()
plt.savefig("Filter_transmission_profile.jpg", dpi=1000)
plt.show()
```
